```python
age_alcohol = pd.crosstab(df['ALCOHOL_CONSUMING'], df['GENDER'])

age_alcohol.plot(kind = 'bar', stacked = True)
```

```python

cols = ['AGE', 'PEER_PRESSURE', 'CHEST_PAIN', 'SWALLOWING_DIFFICULTY', 'SMOKING', 'LUNG_CANCER']

g = sns.PairGrid(df[cols], palette = 'coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap = 'Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
plt.title("PairGrid for selected columns")
plt.show()
```
